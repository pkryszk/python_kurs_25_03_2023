# Podstawy protokołu HTTP


## Wprowadzenie

### Definicja protokołu HTTP

Protokół HTTP (Hypertext Transfer Protocol) to standardowa procedura komunikacyjna, która pozwala na przesyłanie danych w Internecie między klientem (np. przeglądarką internetową) a serwerem.

Protokół HTTP opiera się na modelu klient-serwer, gdzie klient wysyła żądanie HTTP do serwera, a serwer zwraca odpowiedź HTTP z żądanymi informacjami. Protokół HTTP jest powszechnie stosowany w Internecie do przesyłania danych takich jak strony internetowe, grafiki, filmy i inne zasoby.

HTTP działa w warstwie aplikacji i jest często używany w połączeniu z innymi protokołami takimi jak TCP/IP (Transmission Control Protocol/Internet Protocol) lub UDP (User Datagram Protocol).

HTTP korzysta z metod, które określają rodzaj żądania, jakie klient chce wysłać do serwera, oraz kodów odpowiedzi, które serwer zwraca w odpowiedzi na żądanie klienta. Wprowadzenie protokołu HTTPS pozwoliło na dodanie warstwy szyfrowania, co zwiększyło bezpieczeństwo przesyłanych danych.

Historia protokołu HTTP sięga lat 90. XX wieku, a od tego czasu został on nieustannie rozwijany i udoskonalany, dzięki czemu stał się podstawowym narzędziem do przesyłania informacji w Internecie.

### Zastosowanie protokołu HTTP

Protokół HTTP jest powszechnie wykorzystywany w Internecie do przesyłania danych między klientem (np. przeglądarką internetową) a serwerem. Najczęstszym zastosowaniem protokołu HTTP jest przeglądanie stron internetowych, ale może być także wykorzystywany do przesyłania innych zasobów, takich jak obrazy, filmy, dokumenty i wiele innych.

HTTP jest również często stosowany w tworzeniu aplikacji internetowych, takich jak sklepy internetowe, platformy społecznościowe czy aplikacje biznesowe. Protokół ten pozwala na przesyłanie danych w czasie rzeczywistym, co jest szczególnie ważne w przypadku interaktywnych aplikacji internetowych.

HTTP jest również wykorzystywany do integracji różnych systemów, takich jak systemy płatności online, systemy zarządzania treścią czy systemy CRM. Dzięki temu, protokół HTTP stał się podstawowym narzędziem do przesyłania informacji w Internecie i jest niezbędny dla wielu dziedzin życia.

### Historia protokołu HTTP

Protokół HTTP (Hypertext Transfer Protocol) został opracowany w latach 80. XX wieku przez brytyjskiego informatyka Tima Bernersa-Lee. Pierwotnie był to prosty protokół służący do przesyłania dokumentów hipertekstowych (takich jak strony WWW) między komputerami w sieci.

* 1989 - rozpoczęcie prac nad protokołem. 
* 1991 - pierwszy dokument opisujący zachowanie klienta i serwera. Wersja HTTP/0.9 obsługiwała tylko jedną metodę żądania (GET), a serwer odpowiadał tylko treścią dokumentu HTML bez nagłówków. 
* 1992 - HTTP/1.0-draft
* 1996 - HTTP/1.0 wersja wprowadziła wiele ulepszeń, w tym zastosowanie nagłówków HTTP, obsługę nowych metod żądania (POST, HEAD, itp.) oraz kodów odpowiedzi HTTP.
* 1997 - HTTP/1.1 - https://datatracker.ietf.org/doc/html/rfc2068 
* 1999 - HTTP/1.1 - wersja udoskonalona: https://datatracker.ietf.org/doc/html/rfc2616. Znacznie bardziej zaawansowana wersja, która wprowadziła wiele nowych funkcji, takich jak obsługa kompresji, kontrola cache i zapytań pipelining.
* 2015 - HTTP/2.0 https://datatracker.ietf.org/doc/html/rfc7540 Wersja wprowadziła wiele ulepszeń, w tym zmniejszenie opóźnień w przesyłaniu danych i zwiększenie wydajności protokołu.


## Zasada działania protokołu HTTP

Protokół HTTP (Hypertext Transfer Protocol) działa na zasadzie komunikacji klient-serwer, w której klient wysyła zapytania (requests) do serwera, a serwer zwraca odpowiedzi (responses) na te zapytania. Zapytania i odpowiedzi składają się z nagłówków (headers) i treści (body). 

Klient może być przeglądarką internetową, aplikacją mobilną lub innym programem, który chce uzyskać dostęp do zasobów serwera, takich jak pliki HTML, CSS, JavaScript, obrazy, filmy itp. Klient wysyła zapytania do serwera, określając metodę HTTP, adres URL zasobu, a także dodatkowe nagłówki, takie jak nagłówek user-agent, który informuje serwer o rodzaju przeglądarki lub aplikacji, którą używa klient.

Serwer otrzymując zapytanie od klienta, analizuje je i zwraca odpowiedź, która może zawierać żądane zasoby, bądź informacje o błędzie lub innej sytuacji. Odpowiedź składa się z kodu stanu (status code), który informuje klienta o tym, czy zapytanie zostało wykonane poprawnie, oraz z nagłówków i treści.

Protokół HTTP jest protokołem bezstanowym, co oznacza, że serwer nie przechowuje żadnych informacji o poprzednich zapytaniach od klienta, co zwiększa jego skalowalność i wydajność. Aby przesyłać dane między kolejnymi zapytaniami, można wykorzystać mechanizm cookie, który pozwala na przechowywanie danych sesyjnych na stronie klienta.

Współcześnie protokół HTTP rozwija się w kierunku wersji HTTP/3, która wykorzystuje protokół QUIC, co przyspiesza przesyłanie danych w sieci.

### Metody HTTP: 

#### GET, POST, PUT, DELETE

Metody HTTP to sposoby na przesyłanie zapytań (requests) i danych między klientem a serwerem za pomocą protokołu HTTP. Najczęściej stosowanymi metodami są GET, POST, PUT oraz DELETE, które można opisać w następujący sposób:

* GET - metoda służąca do pobierania zasobów (np. stron internetowych, plików), które są dostępne pod określonym adresem URL. Metoda ta przesyła zapytanie do serwera o żądany zasób i otrzymuje odpowiedź w postaci treści tego zasobu. Metoda GET jest bezpieczna, ponieważ nie wprowadza zmian na serwerze.

* POST - metoda służąca do przesyłania danych do serwera w celu wykonania określonej akcji, np. dodania nowego wpisu do bazy danych. Metoda ta przesyła dane wraz z zapytaniem, a serwer zwraca odpowiedź, która może zawierać informacje o sukcesie lub błędzie wykonania akcji.

* PUT - metoda służąca do aktualizacji istniejących zasobów na serwerze, np. edycji pliku lub modyfikacji rekordu w bazie danych. Metoda ta przesyła dane o aktualizowanym zasobie wraz z zapytaniem PUT, a serwer zwraca odpowiedź, która może zawierać informacje o sukcesie lub błędzie aktualizacji.

* DELETE - metoda służąca do usuwania istniejących zasobów na serwerze, np. usuwania plików lub rekordów z bazy danych. Metoda ta przesyła zapytanie o usunięcie zasobu do serwera, a serwer zwraca odpowiedź, która może zawierać informacje o sukcesie lub błędzie usuwania.

Metody HTTP różnią się sposobem przesyłania danych i zapytań między klientem a serwerem, dlatego ważne jest odpowiednie ich stosowanie w zależności od potrzeb i celu przesyłania informacji.

#### Inne metody

Oprócz wymienionych wyżej metod istnieje wiele innych metod, które mogą być używane w zależności od potrzeb i celu przesyłania informacji. Poniżej przedstawiam kilka innych ważnych metod HTTP:

* HEAD - metoda ta jest podobna do metody GET, ale nie pobiera treści zasobu, a jedynie nagłówki HTTP, które zawierają informacje o zasobie, takie jak typ MIME, rozmiar pliku czy datę modyfikacji. Jest to przydatne narzędzie do sprawdzania statusu zasobów i oszczędzania przepustowości sieci.

* OPTIONS - metoda ta służy do pobrania informacji o dostępnych opcjach i metodach dla danego zasobu. Serwer zwraca informacje o tym, jakie metody i nagłówki HTTP są dostępne dla danego zasobu, co może być przydatne dla programistów i automatycznych narzędzi.

* PATCH - metoda ta służy do częściowej aktualizacji istniejącego zasobu. Przesyłane dane zawierają tylko te pola, które mają zostać zmodyfikowane, a nie cały zasób. Jest to przydatne narzędzie do zmiany tylko niektórych informacji w dużych zasobach, co zmniejsza ilość przesyłanych danych i przyspiesza przetwarzanie.

* TRACE - metoda ta służy do zwracania informacji o ścieżce, którą przeszło zapytanie HTTP między klientem a serwerem. Jest to przydatne narzędzie do debugowania i analizowania sieci.

* CONNECT - metoda ta służy do ustanawiania tunelu SSL/TLS do serwera. Jest to przydatne narzędzie do bezpiecznego przesyłania danych między klientem a serwerem.

* PROPFIND - metoda ta służy do pobierania informacji o właściwościach zasobów w hierarchicznym systemie plików, takim jak WebDAV.


### Nagłówki HTTP

Nagłówki HTTP są elementami wiadomości HTTP, które zawierają dodatkowe informacje o żądaniu lub odpowiedzi HTTP. Nagłówki HTTP są używane do przekazywania różnych informacji między klientem a serwerem, takich jak typ zawartości, kodowanie, język, ciasteczka, autoryzacja, kontrola cache, informacje o użytkowniku i wiele innych.

Poniżej przedstawiam kilka nagłówków HTTP, które są szczególnie przydatne:

* Content-Type - nagłówek ten określa typ zawartości wysyłanej lub odbieranej przez serwer. Przykładowe wartości to application/json, text/html, image/png, audio/mpeg itp.

* Content-Length - nagłówek ten określa długość treści wiadomości HTTP. Jest to przydatne narzędzie do kontrolowania przesyłanych danych i zapobiegania przeciążeniom serwera.

* Authorization - nagłówek ten jest używany do przesyłania informacji uwierzytelniających wraz z żądaniem HTTP. Może zawierać tokeny uwierzytelniające, hasła czy klucze API.

* Cookie - nagłówek ten zawiera ciasteczka, które są przechowywane przez klienta i wysyłane wraz z każdym żądaniem HTTP. Są one używane do przechowywania informacji o sesji, ustawieniach użytkownika i innych danych.

* Cache-Control - nagłówek ten określa, jak kontrolować cache w celu zmniejszenia czasu ładowania strony i zmniejszenia obciążenia serwera. Przykładowe wartości to max-age, no-cache, must-revalidate itp.

* User-Agent - nagłówek ten zawiera informacje o przeglądarce lub aplikacji, która wysyła żądanie HTTP. Serwer może użyć tych informacji do dostosowania odpowiedzi do danego klienta.

* Referer - nagłówek ten zawiera adres URL strony, która zawierała link do bieżącej strony. Jest to przydatne narzędzie do śledzenia, skąd przyszedł użytkownik i jakie akcje wykonuje na stronie.

* Accept-Language - nagłówek ten określa preferowany język użytkownika, w którym powinna być dostarczana treść.


### Kod odpowiedzi HTTP

Kody odpowiedzi HTTP to trzycyfrowe liczby, które są przesyłane w nagłówku odpowiedzi HTTP w celu powiadomienia klienta o statusie żądania. Poniżej przedstawiam najważniejsze kody odpowiedzi HTTP:

* 1xx - Informacyjne - oznacza, że żądanie zostało odebrane i jest przetwarzane.

* 2xx - Sukces - oznacza, że żądanie zostało pomyślnie wykonane. Najczęściej spotykanym kodem w tej grupie jest 200 OK.

* 3xx - Przekierowanie - oznacza, że żądanie wymaga przekierowania na inny adres URL. Najczęściej spotykanymi kodami w tej grupie są 301 Moved Permanently, 302 Found i 307 Temporary Redirect.

* 4xx - Błąd klienta - oznacza, że żądanie zawiera błędy i nie może być wykonane. Najczęściej spotykanymi kodami w tej grupie są 400 Bad Request, 401 Unauthorized, 403 Forbidden i 404 Not Found.

* 5xx - Błąd serwera - oznacza, że serwer napotkał problem podczas przetwarzania żądania. Najczęściej spotykanymi kodami w tej grupie są 500 Internal Server Error i 503 Service Unavailable.

Kody odpowiedzi HTTP są ważnym narzędziem do diagnozowania problemów z żądaniami HTTP i pozwalają na szybkie zlokalizowanie przyczyny błędów.

## Charakterystyka wersji protokołu HTTP (10 minut)

Stosowane są obecnie dwie wersje protokołu HTTP - HTTP/1.1 i HTTP/2. Oto krótkie scharakteryzowanie każdej z tych wersji:

* HTTP/1.1:

    - Jest to najbardziej popularna wersja protokołu HTTP stosowana obecnie.
    - Obsługuje wiele żądań w jednym połączeniu za pomocą zapytań pipelining i keep-alive.
    - Wymaga zawsze przesyłania nagłówka Host w każdym żądaniu.
    - Nie obsługuje jednoczesnego przesyłania wielu zasobów przez jedno połączenie.
    - Wymaga osobnego połączenia dla każdego zapytania.

* HTTP/2:

    - Jest to nowsza wersja protokołu HTTP, która wprowadza szereg ulepszeń i optymalizacji.
    - Obsługuje wiele zapytań jednocześnie na jednym połączeniu, co znacznie poprawia wydajność przesyłania.
    - Wykorzystuje jedno połączenie do jednoczesnego przesyłania wielu zasobów, co zwiększa wydajność.
    - Korzysta z technologii multiplexingu, która pozwala na równoczesne przesyłanie wielu zasobów przez jedno połączenie.
    - Wprowadza kompresję nagłówków, co redukuje rozmiar przesyłanych danych i poprawia wydajność.
    - Wymaga użycia protokołu szyfrowanego TLS.
    - Wymaga przesyłania nagłówka ":authority" zamiast nagłówka "Host".


Wcześniejsze wersje nie są obecnie w praktyce wykorzystywane. Ale warto w ramach ciekawostki scharakteryzować wersję HTTP/1.0 z 1996 roku. Oto krótkie scharakteryzowanie tej wersji:

* HTTP/1.0:

    - Każde żądanie i odpowiedź wymaga osobnego połączenia TCP.
    - Nie ma możliwości utrzymania połączenia między serwerem a klientem po przesłaniu odpowiedzi, co wymaga ponownego nawiązania połączenia dla każdego żądania.
    - Wymaga przesyłania nagłówka "Host" w każdym żądaniu, aby określić adres serwera, z którym klient chce się połączyć.
    - Nie obsługuje składni URI względnych, co oznacza, że wszystkie odwołania do zasobów muszą być w formacie bezwzględnym.
    - Nie ma możliwości przesyłania danych w dwóch kierunkach na jednym połączeniu, co ogranicza wydajność przesyłania.
    - Nie obsługuje nagłówków o wartościach złożonych, co utrudnia przesyłanie pewnych rodzajów danych.
    - Wymaga, aby serwer udostępniał informacje o wielkości zasobu przed przesłaniem go do klienta.
    - Wymaga, aby klient określił typ MIME każdego żądania.

HTTP/1.0 jest obecnie rzadko stosowany, ponieważ jego ograniczenia wpłynęły negatywnie na wydajność przesyłania. Wersja ta została zastąpiona przez HTTP/1.1, która wprowadza wiele ulepszeń.

## Przykłady request / response

Oto przykład żądania (request) i odpowiedzi (response) w komunikacji HTTP:

### Request

```yaml
GET /index.html HTTP/1.1
Host: www.example.com
User-Agent: Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:88.0) Gecko/20100101 Firefox/88.0
Accept-Language: en-US,en;q=0.5
Connection: keep-alive
```

W powyższym przykładzie żądanie to zapytanie GET, które ma na celu pobranie pliku "index.html" z serwera o adresie "www.example.com". Nagłówek Host określa adres hosta, a User-Agent informuje serwer o przeglądarce internetowej używanej przez klienta.

### Response

```yaml

HTTP/1.1 200 OK
Date: Thu, 15 Apr 2021 14:45:00 GMT
Server: Apache/2.4.41 (Ubuntu)
Last-Modified: Tue, 13 Apr 2021 10:12:25 GMT
ETag: "20-5c8f3d5b5f800"
Accept-Ranges: bytes
Content-Length: 32
Keep-Alive: timeout=5, max=100
Connection: Keep-Alive
Content-Type: text/html

<html><body>Hello, world!</body></html>
```
W powyższym przykładzie odpowiedź to odpowiedź serwera na żądanie GET. Kod stanu "200 OK" oznacza, że żądanie zostało pomyślnie przetworzone, a w nagłówku Content-Type określony jest typ MIME zwracanego zasobu. W tym przypadku jest to dokument HTML. Zawartość odpowiedzi to prosty dokument HTML zawierający napis "Hello, world!".


## Zabezpieczenia protokołu HTTP (5 minut)


Protokół HTTP nie zapewnia zabezpieczeń na poziomie przesyłanych danych, co oznacza, że przesyłane informacje mogą być łatwo przechwytywane i odczytywane przez osoby trzecie. Aby zwiększyć poziom bezpieczeństwa, można zastosować kilka metod:

1. HTTPS - jest to rozszerzenie protokołu HTTP, które zapewnia szyfrowanie danych przesyłanych między klientem a serwerem. W protokole HTTPS, dane są szyfrowane za pomocą protokołu SSL lub TLS, co uniemożliwia przechwycenie i odczytanie przesyłanych informacji przez osoby trzecie.

2. Szyfrowanie danych - W przypadku, gdy nie jest możliwe użycie protokołu HTTPS, dane można szyfrować przed wysłaniem ich do serwera. W tym celu można użyć protokołu szyfrowania, takiego jak PGP, który zapewni poufność danych przesyłanych w sieci.

3. Autentykacja - Protokół HTTP nie posiada mechanizmów autentykacji, co oznacza, że każdy może twierdzić, że jest kimś innym. Aby zapewnić autentyczność klientów, można zastosować mechanizmy autentykacji, takie jak HTTP Basic Authentication, OAuth, czy JWT.

4. Kontrola dostępu - Protokół HTTP nie zapewnia mechanizmów kontroli dostępu do zasobów, co oznacza, że każdy może uzyskać dostęp do dowolnego zasobu, jeśli zna jego adres URL. Aby zapewnić kontrolę dostępu do zasobów, można zastosować autoryzację na poziomie serwera, np. przez wykorzystanie plików .htaccess, lub przez zastosowanie specjalnych frameworków, takich jak Spring Security czy Django.

5. Weryfikacja integralności danych - Aby zapobiec modyfikowaniu danych podczas ich przesyłania, można użyć mechanizmów weryfikacji integralności danych, takich jak cyfrowe sygnatury, które umożliwiają sprawdzenie, czy dane zostały zmodyfikowane w trakcie przesyłania.

6. Weryfikacja tożsamości serwera - Certyfikaty SSL/TLS to kluczowy element w bezpiecznej komunikacji za pomocą protokołu HTTPS. Są one używane do weryfikacji tożsamości serwera oraz do szyfrowania przesyłanych między nim a klientem danych. <br>Certyfikaty SSL/TLS są wydawane przez zaufane organizacje zwane Autorami Certyfikatów SSL (CA), które po weryfikacji tożsamości właściciela domeny przyznają mu certyfikat SSL/TLS. W certyfikacie znajduje się między innymi nazwa domeny, dla której został wydany, klucz publiczny oraz podpis cyfrowy Autora Certyfikatu SSL.<br>Podczas nawiązywania połączenia HTTPS, klient pobiera certyfikat SSL/TLS serwera i weryfikuje jego autentyczność. Jeśli certyfikat jest prawidłowy, klucz publiczny z certyfikatu jest używany do szyfrowania danych przesyłanych do serwera, a klucz prywatny jest wykorzystywany przez serwer do ich odszyfrowania. W ten sposób zapewniona jest poufność przesyłanych danych oraz autentyczność serwera, z którym nawiązujemy połączenie.

## Podsumowanie 

Najważniejsze elementy protokołu HTTP to: żądania (requests) i odpowiedzi (responses) przesyłane między klientem a serwerem, nagłówki (headers) zawierające informacje o przesyłanych danych oraz status code informujący o wyniku operacji. Protokół HTTP jest używany do przesyłania informacji między klientem a serwerem w internecie.

W praktyce protokół HTTP jest powszechnie używany do przeglądania stron internetowych, pobierania danych z serwerów, przesyłania formularzy i wysyłania wiadomości. W celu zapewnienia bezpieczeństwa przesyłanych danych, zaleca się stosowanie protokołu HTTPS, który zapewnia szyfrowanie przesyłanych danych. Protokół HTTP jest również wykorzystywany w rozwoju aplikacji internetowych, np. w API (interfejsach programowania aplikacji), które umożliwiają komunikację między różnymi serwisami internetowymi.



## Dodatkowe informacje

#### pipelining

Zapytania pipelining to technika optymalizacji przesyłania danych w protokole HTTP. Polega ona na tym, że klient wysyła kolejne zapytania HTTP do serwera, zanim ten zdąży odpowiedzieć na poprzednie. W ten sposób, klient może przyspieszyć przesyłanie danych, ponieważ serwer może zwracać odpowiedzi w kolejności, w jakiej otrzymał zapytania.

Wcześniejsze wersje protokołu HTTP (np. HTTP/1.0) wymagały, aby klient otrzymał odpowiedź na poprzednie żądanie, zanim mógł wysłać kolejne zapytanie. Zapytania pipelining pozwala na jednoczesną obsługę wielu zapytań HTTP, co zwiększa wydajność protokołu i przyspiesza przesyłanie danych.

Warto jednak pamiętać, że zapytania pipelining nie zawsze są skuteczne i mogą wpłynąć na stabilność sieci. Niektóre serwery lub przeglądarki internetowe mogą nie obsługiwać tej techniki, co prowadzi do nieprzewidywalnego zachowania systemu.

#### kontrola cache

Kontrola cache w protokole HTTP polega na zarządzaniu pamięcią podręczną (cache) w przeglądarce internetowej oraz na serwerze. Cache to rodzaj pamięci tymczasowej, w której przechowywane są często używane zasoby, takie jak pliki CSS, JavaScript czy obrazy. Dzięki temu, przeglądarka nie musi pobierać tych zasobów za każdym razem, kiedy użytkownik odwiedza daną stronę internetową, co przyspiesza ładowanie strony.

Kontrola cache pozwala na ustalenie, jak długo dany zasób powinien być przechowywany w cache, a także pozwala na określenie, czy przeglądarka powinna wysyłać zapytanie do serwera w celu pobrania nowszej wersji zasobu. W protokole HTTP wykorzystywane są nagłówki cache-control i expires, które pozwalają na ustawienie czasu ważności danych w cache oraz na kontrolowanie sposobu aktualizacji tych danych.

Dzięki kontroli cache, przeglądarka internetowa może szybciej wyświetlać strony internetowe, a serwer może zmniejszyć obciążenie i przyspieszyć przetwarzanie zapytań, ponieważ nie musi za każdym razem wysyłać tych samych zasobów do klienta.
